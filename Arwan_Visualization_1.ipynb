```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data setup
parameters = [
    "fold1_alert", "fold2_pain",
    "fold3_anxious", "fold4_threatened", "fold5_angry",
    "fold6_stressed", "fold7_defensive", "fold8_stress_release",
    "fold9_friendly", "fold10_affectionate"
]
counts = [100] * 10  # Count value of 100 for all parameters

# Create a DataFrame
data = pd.DataFrame({'Parameter': parameters, 'Count': counts})

# Set up the figure
plt.figure(figsize=(10, 6))

# Create the horizontal bar plot
sns.barplot(y='Parameter', x='Count', data=data, palette='viridis')

# Add labels and title
plt.xlabel('Count', fontsize=12)
plt.ylabel('Parameters', fontsize=12)
plt.title('Horizontal Bar Plot with Viridis Palette', fontsize=14)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np  # Import numpy for random number generation

# Data setup
parameters = [
    "fold1_alert", "fold2_pain",
    "fold3_anxious", "fold4_threatened", "fold5_angry",
    "fold6_stressed", "fold7_defensive", "fold8_stress_release",
    "fold9_friendly", "fold10_affectionate"
]
# Generate random count values between 100 and 150 for each parameter
counts = np.random.randint(100, 151, size=len(parameters))  # Generate 10 random values

# Create a DataFrame
data = pd.DataFrame({'Parameter': parameters, 'Count': counts})

# Set up the figure
plt.figure(figsize=(10, 6))

# Create the vertical bar plot
sns.barplot(x='Parameter', y='Count', data=data, palette='viridis')

# Add labels and title
plt.ylabel('Count', fontsize=12)
plt.xlabel('Parameters', fontsize=12)
plt.title('Bar Plot with Distribution of Classes', fontsize=14)

# Rotate x-ticks for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```
