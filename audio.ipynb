```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion matrix data
confusion_matrix = np.array([[25, 1, 2, 2],
                              [1, 23, 2, 4],
                              [3, 3, 20, 4],
                              [1, 3, 5, 20]])

# Class names
classes = ['Sleep', 'Low Physical Activity', 'Moderate Physical Activity', 'Intense Physical Activity']

# Create heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
epochs = range(1, 21)
training_accuracy = [0.6, 0.65, 0.7, 0.72, 0.75, 0.78, 0.8, 0.82, 0.84, 0.85,
                     0.86, 0.87, 0.87, 0.86, 0.87, 0.87, 0.86, 0.87, 0.87, 0.87]
validation_accuracy = [0.58, 0.62, 0.65, 0.68, 0.70, 0.73, 0.76, 0.78, 0.80, 0.82,
                       0.83, 0.84, 0.84, 0.83, 0.84, 0.84, 0.83, 0.84, 0.85, 0.85]
training_loss = [0.6, 0.55, 0.5, 0.48, 0.45, 0.4, 0.35, 0.32, 0.3, 0.28,
                 0.26, 0.25, 0.25, 0.26, 0.25, 0.25, 0.26, 0.25, 0.25, 0.25]
validation_loss = [0.65, 0.62, 0.6, 0.58, 0.55, 0.54, 0.52, 0.5, 0.48, 0.47,
                   0.45, 0.42, 0.43, 0.42, 0.4, 0.4, 0.4, 0.39, 0.38, 0.37]

# Create figure and axis
fig, ax1 = plt.subplots(2, 1, figsize=(10, 10))

# Plotting accuracy
ax1[0].plot(epochs, training_accuracy, label='Training Accuracy', marker='o')
ax1[0].plot(epochs, validation_accuracy, label='Validation Accuracy', marker='o')
ax1[0].set_title('Learning Curve')
ax1[0].set_ylabel('Accuracy')
ax1[0].set_xticks(epochs)
ax1[0].legend()
ax1[0].grid()

# Plotting loss
ax1[1].plot(epochs, training_loss, label='Training Loss', marker='o', color='red')
ax1[1].plot(epochs, validation_loss, label='Validation Loss', marker='o', color='orange')
ax1[1].set_title('Error Curve')
ax1[1].set_ylabel('Loss')
ax1[1].set_xlabel('Epochs')
ax1[1].set_xticks(epochs)
ax1[1].legend()
ax1[1].grid()

# Show plot
plt.tight_layout()
plt.show()
```
