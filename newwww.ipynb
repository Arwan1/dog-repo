```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the class labels for the three breeds
class_names = [
    "german-shepard_alert", "german-shepard_suspicious", "german-shepard_anxious", 
    "german-shepard_threatened", "german-shepard_angry", "german-shepard_peace", 
    "german-shepard_stressed", "german-shepard_respect", "german-shepard_needs_space", 
    "german-shepard_stalking", "german-shepard_stress_released", "german-shepard_relaxed", 
    "german-shepard_polite", "german-shepard_friendly", "german-shepard_lovebug",
    "ShihTzu_alert", "ShihTzu_suspicious", "ShihTzu_anxious", 
    "ShihTzu_threatened", "ShihTzu_angry", "ShihTzu_peace", 
    "ShihTzu_stressed", "ShihTzu_respect", "ShihTzu_space", 
    "ShihTzu_stalking", "ShihTzu_stress_released", "ShihTzu_relaxed", 
    "ShihTzu_polite", "ShihTzu_friendly", "ShihTzu_lovebug",
    "CockerSpaniel_alert", "CockerSpaniel_suspicious", "CockerSpaniel_anxious", 
    "CockerSpaniel_threatened", "CockerSpaniel_angry", "CockerSpaniel_peace", 
    "CockerSpaniel_stressed", "CockerSpaniel_respect", "CockerSpaniel_needs_space", 
    "CockerSpaniel_stalking", "CockerSpaniel_stress_released", "CockerSpaniel_relaxed", 
    "CockerSpaniel_polite", "CockerSpaniel_friendly", "CockerSpaniel_lovebug"
]

# Parameters
num_classes = len(class_names)
samples_per_class = 300
accuracy = 0.81

# Initialize the confusion matrix
confusion_matrix = np.zeros((num_classes, num_classes), dtype=int)

# Populate the confusion matrix with variable correct predictions around the target accuracy
for i in range(num_classes):
    correct_predictions = int(np.random.normal(samples_per_class * accuracy, 5))  # adding variability
    incorrect_predictions = samples_per_class - correct_predictions
    
    # Fill the diagonal with correct predictions
    confusion_matrix[i, i] = correct_predictions

    # Distribute incorrect predictions randomly across off-diagonal cells
    incorrect_indices = [j for j in range(num_classes) if j != i]
    misclassifications = np.random.choice(incorrect_indices, incorrect_predictions, replace=True)
    for idx in misclassifications:
        confusion_matrix[i, idx] += 1

# Plot the confusion matrix
plt.figure(figsize=(12, 10))
sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Greens", 
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Simulated Confusion Matrix for Dog Emotion Classification ")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
